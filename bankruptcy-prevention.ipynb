# 2. KNN model

```python
from sklearn.neighbors import KNeighborsClassifier as KNC
import warnings
warnings.filterwarnings('ignore')
```

### To choose k value

```python
import math
math.sqrt(len(y_test))
```

Here we are choosing the k value to be  7 (choosing odd value)

Define the model KNN and fit model

```python
KNN_classifier = KNC(n_neighbors =7, p = 2, metric = 'euclidean')
```

```python
KNN_classifier.fit(x_train, y_train)
```

#### Predict the Test set results

```python
y_pred = KNN_classifier.predict(x_test)
y_pred
```

```python
y_pred_df1=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
y_pred_df1
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
y_pred_probs = KNN_classifier.predict_proba(x_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_probs)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
auc = roc_auc_score(y_test, y_pred_probs)
auc
```

```python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_probs)

plt.figure(figsize = (5,4))
plt.plot([0, 1], [0.5, 0.5],'k--')
plt.plot(recall, precision, label = 'Knn')
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('Knn(n_neighbors = 7) PRC curve')
plt.show()
```
